```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv("Smartphone_Usage_And_Addiction_Analysis_7500_Rows.csv")

# 컬럼 한글 이름 매핑
df_kr = df.rename(columns={
    'daily_screen_time_hours': '일일 사용시간',
    'social_media_hours': 'SNS 사용시간',
    'gaming_hours': '게임 사용시간',
    'sleep_hours': '수면 시간',
    'notifications_per_day': '알림 수',
    'app_opens_per_day': '앱 실행 횟수',
    'weekend_screen_time': '주말 사용시간',
    'addicted_label': '중독 여부'
})

cols = [
    '일일 사용시간',
    'SNS 사용시간',
    '게임 사용시간',
    '수면 시간',
    '알림 수',
    '앱 실행 횟수',
    '주말 사용시간',
    '중독 여부'
]

corr = df_kr[cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)

plt.title("스마트폰 사용 데이터 상관관계 히트맵")
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()

plt.savefig("heatmap.png", dpi=300, bbox_inches='tight')
```
